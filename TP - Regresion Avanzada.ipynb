```python
import pandas as pd

grasacerdos= pd.read_excel("grasacerdos.xlsx")

grasacerdos.info()
```

```python
import matplotlib.pyplot as plt

x= grasacerdos["PV"]
y= grasacerdos["EGD"]

plt.scatter(x, y, marker='o')

# Personalizar el gráfico
plt.title('Diagrama de Dispersión del Peso vivo (PV) y el Espesor de grasa dorsal (EGD)')
plt.xlabel('Peso vivo')
plt.ylabel('Espesor de grasa dorsal')
plt.grid(False)
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np

r = np.corrcoef(x, y)[0, 1]
print("r:", r)
```

Interpretación:

* Este valor indica una correlación positiva débil entre el peso y el espesor de grasa dorsal.

* Aunque hay una tendencia a que ambos aumenten juntos, la relación no es fuerte.

* Esto podría sugerir que otras variables también influyen en el espesor de grasa, o que la relación no es completamente lineal.
